## Q1: What are the most popular boroughs and zones?

```python
query_boroughs_zones = """
SELECT
    dl.zone,
    dl.borough,
    COUNT(ft.trip_id) AS number_of_trips
FROM 
    public.dim_location dl
JOIN 
    public.fct_trips ft ON dl.location_id = ft.start_location_id
GROUP BY 
    dl.borough, dl.zone
ORDER BY 
    number_of_trips DESC;
"""
df_boroughs_zones = pd.read_sql(query_boroughs_zones, url)
```

```python
df_boroughs_zones
```

```python
query_boroughs = """
SELECT 
    dl.borough,
    ft.type_description,
    COUNT(ft.trip_id) AS number_of_trips
FROM 
    public.dim_location dl
JOIN 
    public.fct_trips ft ON dl.location_id = ft.start_location_id
GROUP BY 
    dl.borough, ft.type_description
ORDER BY 
    ft.type_description, number_of_trips DESC;
"""
df_boroughs = pd.read_sql(query_boroughs, url)
```

```python
df_boroughs.head(20)
# Group by 'type_description', and find the borough with the maximum 'number_of_trips' for each type
most_popular_by_type = df_boroughs.groupby('type_description').apply(lambda x: x[x['number_of_trips'] == x['number_of_trips'].max()])
most_popular_by_type
```

### ZONES

```python
query_zones = """
SELECT 
    dl.zone,
    ft.type_description,
    COUNT(ft.trip_id) AS number_of_trips
FROM 
    public.dim_location dl
JOIN 
    public.fct_trips ft ON dl.location_id = ft.start_location_id
GROUP BY 
    dl.zone, ft.type_description
ORDER BY 
    dl.zone, number_of_trips DESC;
"""

df_zones = pd.read_sql(query_zones, url)
```

## Q2: How does the overall pattern look like over a year?

```python
query_year = """
SELECT 
    dd.month,
    ft.type_description,
    COUNT(ft.trip_id) AS number_of_trips
FROM 
    fct_trips ft
JOIN 
    dim_date dd ON ft.start_date_id = dd.date
GROUP BY 
    dd.month, ft.type_description
ORDER BY 
    dd.month, number_of_trips DESC;
"""
df_year = pd.read_sql(query_year, url)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Creating the stacked bar plot
plt.figure(figsize=(14, 8))
sns.barplot(data=df_year, x='month', y='number_of_trips', hue='type_description', palette='viridis')

plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.title('Number of Trips per Month by Type')
plt.xticks(rotation=45)  # Rotate the month labels for better readability
plt.legend(title='Trip Type')
plt.tight_layout()  # Adjust layout to make room for the rotated month labels
```

## Q3: How does the hourly pattern look like over a day?
### Also: Does time of the day affect the taxi ridership?

```python
query_hour = """
SELECT
    ft.type_description,
    dt.hour,
    COUNT(ft.trip_id) AS number_of_trips
FROM 
    fct_trips ft
JOIN 
    dim_time dt ON ft.start_time_id = dt.time
GROUP BY 
    dt.hour, ft.type_description
ORDER BY 
    number_of_trips DESC;
"""
```

## Q4: How does the hourly pattern look like over a week?

```python
query_hour_weekly = """
SELECT 
    dd.week_day,
    dt.hour,
    COUNT(ft.trip_id) AS number_of_trips
FROM 
    fct_trips ft
JOIN 
    dim_time dt ON ft.start_time_id = dt.time
JOIN
    dim_date dd ON ft.start_date_id = dd.date
GROUP BY 
    dd.week_day, dt.hour
ORDER BY 
    dd.week_day, dt.hour;
"""
```

```python
df_hour_weekly = pd.read_sql(query_hour_weekly, url)
```

```python
import seaborn as sns

# Creating a pivot table for the heat map correctly
pivot_table = df_hour_weekly.pivot_table(index='hour', columns='week_day', values='number_of_trips', aggfunc='sum')

# Plotting the heat map
plt.figure(figsize=(12, 7))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
plt.title('Hourly Trips Pattern Over a Week')
plt.xlabel('Hour of Day')
plt.ylabel('Day of the Week')
plt.show()
```

## Q5: What are the duration and distance of the trips?

```python
query_duration_distance = """
SELECT 
    AVG(duration_sec) AS average_duration_seconds,
    MIN(duration_sec) AS minimum_duration_seconds,
    MAX(duration_sec) AS maximum_duration_seconds,
    AVG(distance_km) AS average_distance_km,
    MIN(distance_km) AS minimum_distance_km,
    MAX(distance_km) AS maximum_distance_km
FROM 
    fct_trips
WHERE 
    type_description = 'taxi'; 
"""
df_duration_distance = pd.read_sql_query(query_duration_distance, url)
```

```python
average_duration_min = df_duration_distance['average_duration_seconds'][0] / 60  
minimum_duration_min = df_duration_distance['minimum_duration_seconds'][0] / 60
maximum_duration_min = df_duration_distance['maximum_duration_seconds'][0] / 60
average_distance = df_duration_distance['average_distance_km'][0]
minimum_distance = df_duration_distance['minimum_distance_km'][0]
maximum_distance = df_duration_distance['maximum_distance_km'][0]

# Print the results
print(f"Average duration: {average_duration_min:.2f} min")
print(f"Minimum duration: {minimum_duration_min:.2f} min")
print(f"Maximum duration: {maximum_duration_min:.2f} min")
print(f"\n\nAverage distance: {average_distance:.2f} km")
print(f"Minimum distance: {minimum_distance:.2f} km")
print(f"Maximum distance: {maximum_distance:.2f} km")
```

```python
query_1 = """
select
    CASE WHEN type_description IN ('classic_bike', 'docked_bike' , 'electric_bike') THEN 'bike' ELSE type_description END AS type_group,
    dd.week_day,
    AVG(duration_sec)/60 as avg_duration,
    AVG(distance_km) as avg_distance
FROM fct_trips ft
INNER JOIN dim_date dd
    ON ft.start_date_id = dd.date
WHERE dd.year = 2023
GROUP BY CASE WHEN type_description IN ('classic_bike', 'docked_bike' , 'electric_bike') THEN 'bike' ELSE type_description END, week_day
"""
df_1 = pd.read_sql(query_1, url)
df_1
```

## Q6: How much is the tip?
### Also: How does tipping rate change according to location?
### Also: How does tipping rate change according to the hours of the day (peak hours)? 

```python
query_tip_location = """
SELECT 
    dl.borough,
    AVG(ft.tip_amount_dol / NULLIF(ft.total_amount_dol, 0)) AS average_tipping_rate
FROM 
    fct_trips ft
JOIN 
    dim_location dl ON ft.start_location_id = dl.location_id
WHERE 
    ft.type_description = 'taxi'
GROUP BY 
    dl.borough
ORDER BY 
    average_tipping_rate DESC;
"""
df_tip_location = pd.read_sql_query(query_tip_location, url)
```

```python
# Plotting average tipping rate by location
plt.figure(figsize=(10, 6))
sns.barplot(x='borough', y='average_tipping_rate', data=df_tip_location)
plt.title('Average Tip by Location USD')
plt.xlabel('Location')
plt.ylabel('Average Tip')
plt.xticks(rotation=45)
plt.show()
```

```python
# falta um filtro sobre apenas as viajens que têm tip
query_tip_hours = """
SELECT 
    dt.hour,
    AVG(ft.tip_amount_dol/ NULLIF(ft.total_amount_dol, 0)) AS average_tipping_rate
FROM 
    fct_trips ft
JOIN 
    dim_time dt ON ft.start_time_id = dt.time
WHERE 
    ft.type_description = 'taxi'
GROUP BY 
    dt.hour
ORDER BY 
    dt.hour;
"""
df_tip_hours = pd.read_sql_query(query_tip_hours, url)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='hour', y='average_tipping_rate', data=df_tip_hours, color= "gold")
plt.title('Average Tip by Hour of Day USD')
plt.xlabel('Hour of Day')
plt.ylabel('Average Tipping')
plt.grid(True)
plt.show()
```

## Q7: Do people travel in group/share rides or prefer to ride alone? (only for taxis)
### Also: Does this pattern change during the year (e.g., holiday seasons)?

```python
query_passenger = """
SELECT 
    passengers_total,
    COUNT(trip_id) AS number_of_trips
FROM 
    fct_trips
WHERE 
    type_description = 'taxi' 
GROUP BY 
    passengers_total
ORDER BY 
    number_of_trips DESC;

"""
df_passenger = pd.read_sql_query(query_passenger, url)
```

```python
# Calculate the total number of trips
total_trips = df_passenger['number_of_trips'].sum()

# Calculate the percentage of the total trips for each passenger count
df_passenger['percentage_of_total_trips'] = (df_passenger['number_of_trips'] / total_trips) * 100
df_passenger
```

```python
query_passenger_year = """
SELECT 
    dd.month,
    AVG(ft.passengers_total) AS avg_passengers
FROM 
    fct_trips ft
JOIN 
    dim_date dd ON ft.start_date_id = dd.date
WHERE 
    ft.type_description = 'taxi'
GROUP BY 
    dd.month
ORDER BY 
    dd.month;
"""

df_passenger_year = pd.read_sql_query(query_passenger_year, url)
```

## Q8: What are the most frequent pickup and drop-off places?
### Also: How do they relate to specific times of the year?

```python
query_pickups = """
SELECT 
    dl.borough AS pickup_location,
    COUNT(ft.trip_id) AS number_of_pickups
FROM 
    fct_trips ft
JOIN 
    dim_location dl ON ft.start_location_id = dl.location_id
GROUP BY 
    dl.borough
ORDER BY 
    number_of_pickups DESC
LIMIT 10;
"""

query_dropoffs = """
SELECT 
    dl.borough AS dropoff_location,
    COUNT(ft.trip_id) AS number_of_dropoffs
FROM 
    fct_trips ft
JOIN 
    dim_location dl ON ft.end_location_id = dl.location_id
GROUP BY 
    dl.borough
ORDER BY 
    number_of_dropoffs DESC
LIMIT 10;
"""

df_pickups = pd.read_sql_query(query_pickups, url)
df_dropoffs = pd.read_sql_query(query_dropoffs, url)

# Display results
print("Most Frequent Pickup Locations:")
print(df_pickups)
print("\nMost Frequent Drop-off Locations:")
print(df_dropoffs)
```

## Q9: Does the payment type utilized by the passenger affect the total cost of the trip?

```python
# Execute the query
query_payment_cost = """
SELECT 
    payment_description, 
    AVG(total_amount_dol) AS average_total_cost,
    MIN(total_amount_dol) AS minimum_total_cost,
    MAX(total_amount_dol) AS maximum_total_cost,
    COUNT(trip_id) AS number_of_trips
FROM 
    fct_trips
GROUP BY 
    payment_description
ORDER BY 
    average_total_cost DESC;
"""

df_payment_cost = pd.read_sql_query(query_payment_cost, url)
```

```python
# Display results
print(df_payment_cost)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='payment_description', y='average_total_cost', data=df_payment_cost)
plt.title('Average Total Cost by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Average Total Cost ($)')
plt.xticks(rotation=45)
plt.show()
```
