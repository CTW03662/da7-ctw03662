## Removal of Nulls

```python
cars.dropna(); # Remover todas as linhas que tenham pelos menos um valor nulo
cars.dropna(how = "all"); # Remover linhas em que TODOS os valores sejam null
cars.dropna(axis = 1); # Dropna aplicado às colunas (remove colunas com valores null em vez de linhas)
cars.dropna(subset = ["Market Category", "Engine Cylinders"]); # selecionando apenas colunas que nos interessam
```

# REPLACE NULL VALUES

```python
cars["Market Category"].fillna("Other")
```

```python
cars["Engine HP"].fillna(cars["Engine HP"].mean())
```

```python
# para variaveis categoricas
cars["Market Category"].fillna(cars["Market Category"].mode())
```

# DUPLICATES

```python
cars.drop_duplicates();
```

```python
cars.duplicated() # indica linhas que têm duplicados
cars[cars.duplicated()];
```

```python
cars.drop_duplicates(subset =["Model"]); # apaga duplicados da coluna Model
```

```python
cars.drop_duplicates(subset = ["Model"], keep = "last"); # fica com a última ocurrência
cars.drop_duplicates(subset = ["Model"], keep = False); # as linhas que estavam duplicadas desaparecem todas
```

# AGGREGATIONS

```python
cars_by_brand = cars.groupby("Make")
```

```python
cars_by_brand.agg({
    "MSRP" : "mean",
    "Popularity" : "max"
})
```

```python
query = """
select 
    Make,
    avg(MSRP) as mean_price,
    max(popularity) as max_popularity
from cars
group by Make
order by Make asc
"""

duckdb.query(query).to_df()
```

# SORT VALUES

```python
cars.sort_values("Year", ascending = False);
cars.sort_values(["Year", "Make"], ascending = [False, True]);
duckdb.query("select * from cars order by Year desc, Make asc").to_df();
```
